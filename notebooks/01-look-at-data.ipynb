```python
import sys
import random

from PIL import Image
from IPython.display import display
import numpy as np

sys.path.append("../")
from src import create_hf_dataset
```

```python
def resize_image_aspect_ratio(image, target_size):
    width, height = image.size
    aspect_ratio = width / height

    if width < height:
        new_width = target_size
        new_height = round(new_width / aspect_ratio)
    else:
        new_height = target_size
        new_width = round(new_height * aspect_ratio)

    resized_image = image.resize((new_width, new_height))

    return resized_image
```

# Deep Fashion Dataset

```python
data_dir = "../data/deep_fashion"
hf_datset = create_hf_dataset(data_dir)
```

```python
example_indices = np.random.randint(0, len(hf_datset), 4).tolist()

dash_line = "-".join(" " for _ in range(100))

for i, index in enumerate(example_indices):
    image = Image.open(hf_datset[index]["image_path"])
    image = resize_image_aspect_ratio(image, 400)
    text = random.choice(hf_datset[index]["texts"])
    
    print(dash_line)
    print(f"Example {i+1}")
    print(dash_line)
    print("IMAGE:")
    display(image)
    print(dash_line)
    print("TEXT:")
    print(text)
    print(dash_line)
    print()
```
