```python
import sys

import torch

sys.path.append("../../")
from src import Generator, AutoDiffusionPipeline
```

```python
model_log_dir = "../../model_logs/experiment2/deep_fashion"
```

```python
restore_version = "train_240404194859"
```

```python
lora_generator = Generator.for_pretrained_lora(model_log_dir, restore_version, dtype=torch.float16, enable_xformers=True, refiner_name="stabilityai/stable-diffusion-xl-refiner-1.0")
base_generator = Generator.for_base_model("stabilityai/stable-diffusion-xl-base-1.0", dtype=torch.float16, enable_xformers=True, refiner_name="stabilityai/stable-diffusion-xl-refiner-1.0")
lora_generator.to("cuda:0")
base_generator.to("cuda:1")
```

# Sample 1

```python
prompt =  "His tank top has no sleeves, cotton fabric and pure color patterns. The neckline of it is round."
seed = 2
best_of_n = 1
guidance_scale = 5
```

```python
lora_images = lora_generator(prompt, best_of_n, seed, refine_frac=0.2, height=512, width=360, guidance_scale=guidance_scale)
lora_images[0]
```

```python
base_images = base_generator(prompt, best_of_n, seed, refine_frac=0.2, height=512, width=360, guidance_scale=guidance_scale, negative_prompt="drawing, cartoon, painting, illustration, 3d, render, cgi")
base_images[0]
```
